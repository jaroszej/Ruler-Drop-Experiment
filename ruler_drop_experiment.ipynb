# Ruler Drop Reaction Time Experiment

This experiment aims to measure reaction times for catching a falling ruler. Follow these steps to perform the experiment with a friend.

---

## Drop Instructions

1. **Positioning**:
   - Have your friend hold a ruler vertically above your open hand, with the zero mark (or starting point) aligned just above your thumb and index finger.
   - Keep your hand open but positioned so that the ruler is directly above it without any space between your hand and the ruler. Avoid touching the ruler to ensure the reaction is unprompted.

2. **Prepare to Catch**:
   - Relax your hand but be ready to close it to catch the ruler as soon as it starts falling.

3. **The Drop**:
   - Without warning, your friend should release the ruler, letting it drop straight down at the speed of gravity. They should not push or throw the ruler.
   - As quickly as possible, close your hand to catch the ruler. Once closed, do not move your hand!

4. **Measure the Drop**:
   - After catching the ruler, look at the measurement on the ruler where you caught it. This distance represents the drop length in inches.

5. **Record the Distance**:
   - Write down the distance in inches from the starting point to where you caught the ruler. This mark will be at the top of your closed hand on the ruler. Repeat this process for several trials to gather multiple measurements. The recommended amount is 30 trials.

````python
import matplotlib.pyplot as plt
import json
import csv
import time
import itertools
import statistics as stat
from datetime import datetime
from scipy.stats import iqr


def plot(subject_name=None, subject_list=None, index_range=None, date=None, date_range=None,
         title="Reaction Times", xlabel="Index", ylabel="Reaction Time (ms)",
         show_mean=False, show_median=False, show_std_band=False, show_best_fit=True):
    """
    Plot reaction times with filtering options and visual enhancements for clarity.
    Parameters allow control over the display of mean, median, standard deviation band, and best-fit line.

    Parameters:
    subject_name (str, optional): The name of a single subject to filter by.
    subject_list (list, optional): A list of subject names to filter by.
    index_range (tuple, optional): A tuple containing the start and end index (inclusive) to filter by.
    date (str, optional): A specific date in "day-month-year" format to filter by.
    date_range (tuple, optional): A tuple containing start and end dates in "day-month-year" format to filter by.
    title (str): The base title for the plot.
    xlabel (str): The label for the x-axis.
    ylabel (str): The label for the y-axis.
    show_mean (bool): If True, show the mean line for each subject's data.
    show_median (bool): If True, show the median line for each subject's data.
    show_std_band (bool): If True, show a shaded standard deviation band around the mean.
    show_best_fit (bool): If True, show a line of best fit for each subject's data.
    """
    try:
        with open("reaction_times.json", "r") as file:
            data = json.load(file)

    except FileNotFoundError:
        print("No records found. The JSON file does not exist.")
        return

    try:
        filtered_data = data
        if subject_name:
            filtered_data = [record for record in filtered_data if record["subject_name"] == subject_name]
        if subject_list:
            filtered_data = [record for record in filtered_data if record["subject_name"] in subject_list]
        if index_range:
            start_idx, end_idx = index_range
            filtered_data = [record for record in filtered_data if start_idx <= record["index"] <= end_idx]
        if date:
            date = datetime.strptime(date, "%d-%m-%Y")
            filtered_data = [record for record in filtered_data if datetime.strptime(record["date"], "%d-%m-%Y") == date]
        if date_range:
            start_date = datetime.strptime(date_range[0], "%d-%m-%Y")
            end_date = datetime.strptime(date_range[1], "%d-%m-%Y")
            filtered_data = [record for record in filtered_data if start_date <= datetime.strptime(record["date"], "%d-%m-%Y") <= end_date]

        if not filtered_data:
            print("No data available for the specified filter criteria.")
            return

        dates = [datetime.strptime(record["date"], "%d-%m-%Y") for record in filtered_data]
        if len(set(dates)) > 1:
            min_date = min(dates).strftime("%d-%m-%Y")
            max_date = max(dates).strftime("%d-%m-%Y")
            title = f"{title} ({min_date} to {max_date})"

        subjects = set(record["subject_name"] for record in filtered_data)
        plt.figure(figsize=(10, 6))

        colors = plt.get_cmap("tab10").colors
        markers = ['o', 's', 'D', '^', 'v', '<', '>', 'p', '*', 'h']
        color_cycle = itertools.cycle(colors)
        marker_cycle = itertools.cycle(markers)

        for subject in subjects:
            subject_data = [record for record in filtered_data if record["subject_name"] == subject]
            normalized_indices = list(range(len(subject_data)))
            reaction_times = [record["reaction_time_ms"] for record in subject_data]
            color = next(color_cycle)
            marker = next(marker_cycle)
            plt.plot(normalized_indices, reaction_times, marker=marker, linestyle='', color=color,
                     label=f"{subject} (n = {len(reaction_times)})", markersize=5)

            if show_mean:
                mean_rt = stat.mean(reaction_times)
                plt.axhline(mean_rt, color=color, linestyle='-', linewidth=1, alpha=0.5, label=f"{subject} Mean")

            if show_median:
                median_rt = stat.median(reaction_times)
                plt.axhline(median_rt, color=color, linestyle='--', linewidth=1, alpha=0.5, label=f"{subject} Median")

            if show_std_band and len(reaction_times) > 1:
                std_dev = stat.stdev(reaction_times)
                mean_rt = stat.mean(reaction_times)
                plt.fill_between(normalized_indices, mean_rt - std_dev, mean_rt + std_dev, color=color, alpha=0.2)

            if show_best_fit and len(normalized_indices) > 1:
                m, b = np.polyfit(normalized_indices, reaction_times, 1)
                best_fit = np.polyval([m, b], normalized_indices)
                plt.plot(normalized_indices, best_fit, '--', color=color, alpha=0.6, label=f"{subject} Best Fit")

        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.title(title)

        if len(subjects) > 1:
            plt.legend(loc="upper right")
        else:
            plt.legend()

        plt.grid(True, zorder=0)
        plt.show()

    except Exception as e:
        print(f"An unexpected error occurred: {e}")


def record_reaction_time(distance_in_inches):
    """
    Calculate and record the reaction time in milliseconds for a ruler falling a given distance in inches.

    Parameters:
    distance_in_inches (float): The distance the ruler fell, in inches.

    Returns:
    float: The reaction time in milliseconds.
    """
    try:
        g = 9.8
        inches_to_meters = 0.0254
        d_meters = distance_in_inches * inches_to_meters
        t_seconds = (2 * d_meters / g) ** 0.5
        reaction_time_ms = t_seconds * 1000

        return round(reaction_time_ms, 4)

    except TypeError:
        print("Error: 'distance_in_inches' must be a numeric value.")
        return None


def import_from_csv(filename="reaction_times.csv"):
    """
    Imports reaction time records from a CSV file and creates a JSON file.

    Parameters:
    filename (str): The name of the CSV file to import from.
    """
    try:
        with open(filename, mode="r") as csvfile:
            reader = csv.DictReader(csvfile)
            data = []
            for row in reader:
                try:
                    record = {
                        "index": int(row["Index"]),
                        "subject_name": row["Subject Name"],
                        "date": row["Date"],
                        "distance_in_inches": float(row["Distance Fallen (inches)"]),
                        "reaction_time_ms": float(row["Reaction Time (ms)"])
                    }
                    data.append(record)
                except (KeyError, ValueError):
                    print("Error in CSV format or data. Ensure headers and data types are correct.")
                    return

        with open("reaction_times.json", "w") as jsonfile:
            json.dump(data, jsonfile, indent=4)

        print(f"Data successfully imported from {filename} to reaction_times.json.")

    except FileNotFoundError:
        print(f"The file {filename} does not exist.")


def export_to_csv(filename="reaction_times.csv"):
    """
    Exports all reaction time records to a CSV file and creates a backup JSON file with a timestamp.

    Parameters:
    filename (str): The name of the CSV file to export to.
    """
    try:
        with open("reaction_times.json", "r") as file:
            data = json.load(file)

        headers = ["Index", "Subject Name", "Date", "Distance Fallen (inches)", "Reaction Time (ms)"]
        with open(filename, mode="w", newline="") as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(headers)
            for record in data:
                writer.writerow([
                    record["index"],
                    record["subject_name"],
                    record["date"],
                    record["distance_in_inches"],
                    record["reaction_time_ms"]
                ])

        # Create a timestamp
        timestamp = int(time.time())
        backup_filename = f"reaction_times_{timestamp}.json"
        with open(backup_filename, "w") as backup_file:
            json.dump(data, backup_file, indent=4)

        print(f"Records have been exported to {filename} and backed up as {backup_filename}.")

    except FileNotFoundError:
        print("No records found to export.")


def view_record(index=None, end_index=None):
    """
    Display records from the JSON file. If an index is provided, return the specified record.
    If a range is provided (index and end_index), return records within that range.
    If no index is provided, return all records.

    Parameters:
    index (int, optional): The index of a specific record to view.
    end_index (int, optional): The end of an index range to view multiple records.
    """
    try:
        with open("reaction_times.json", "r") as file:
            data = json.load(file)

        if not data:
            print("No records found.")
            return

        if index is not None and end_index is not None:
            filtered_records = [record for record in data if index <= record["index"] <= end_index]
        elif index is not None:
            filtered_records = [record for record in data if record["index"] == index]
        else:
            filtered_records = data

        if not filtered_records:
            print("No records found for the specified index or range.")
        else:
            print("Records:\n")
            for i, record in enumerate(filtered_records):
                print(f"Index: {record['index']}\n"
                      f"Subject: {record['subject_name']}\n"
                      f"Date: {record['date']}\n"
                      f"Distance Fallen: {record['distance_in_inches']} inches\n"
                      f"Reaction Time: {record['reaction_time_ms']} ms\n")
                if i < len(filtered_records) - 1:
                    print("-" * 40)  # Separator line
            return

    except FileNotFoundError:
        print("No records found. The JSON file does not exist.")


def search_records(subject_name=None, date=None):
    """
    Search for records by subject name or date.

    Parameters:
    subject_name (str, optional): The name of the subject to search for.
    date (str, optional): The date to search for in "day-month-year" format.
    """
    try:
        # Load existing data from the JSON file
        with open("reaction_times.json", "r") as file:
            data = json.load(file)

        # Filter records based on search criteria
        filtered_records = [
            record for record in data
            if (subject_name is None or record["subject_name"] == subject_name) and
               (date is None or record["date"] == date)
        ]

        if not filtered_records:
            print("No records found for the specified criteria.")
        else:
            print("Search Results:\n")
            for i, record in enumerate(filtered_records):
                print(f"Index: {record['index']}\n"
                      f"Subject: {record['subject_name']}\n"
                      f"Date: {record['date']}\n"
                      f"Distance Fallen: {record['distance_in_inches']} inches\n"
                      f"Reaction Time: {record['reaction_time_ms']} ms\n")
                if i < len(filtered_records) - 1:
                    print("-" * 40)  # Separator line
            return

    except FileNotFoundError:
        print("No records found. The JSON file does not exist.")


def clear_records(index=None):
    """
    Clear records in the JSON file. If an index is provided, only the record with that index will be removed.
    If no index is provided, the entire file will be cleared.

    Parameters:
    index (int, optional): The index of the record to delete. If None, all records are cleared.
    """
    try:
        with open("reaction_times.json", "r") as file:
            data = json.load(file)

        if index is not None:
            data = [record for record in data if record["index"] != index]
            # Update indices for remaining records to maintain consistency
            for i, record in enumerate(data):
                record["index"] = i + 1
            print(f"Record with index {index} has been cleared.")
        # If no index is provided, clear all records
        else:
            data = []
            print("All records have been cleared.")

        with open("reaction_times.json", "w") as file:
            json.dump(data, file, indent=4)

    except FileNotFoundError:
        print("No records found to clear.")


def stats(subject_name=None, subject_list=None, date=None, date_range=None, index_range=None):
    """
    Calculate comprehensive statistics (average, minimum, maximum, median, standard deviation, count,
    range, interquartile range, 25th percentile, 75th percentile, mode, and variance) for reaction times.
    Allows filtering by subject, list of subjects, specific date, date range, or index range.

    Parameters:
    subject_name (str, optional): Filter by a single subject's name.
    subject_list (list, optional): Filter by a list of subject names.
    date (str, optional): Specific date in "day-month-year" format to filter by.
    date_range (tuple, optional): Tuple with start and end dates in "day-month-year" format to filter by.
    index_range (tuple, optional): Tuple containing start and end indices to filter by.

    Returns:
    list: A list containing the calculated statistics.
    """
    try:
        with open("reaction_times.json", "r") as file:
            data = json.load(file)

        filtered_data = data

        if subject_name:
            filtered_data = [record for record in filtered_data if record["subject_name"] == subject_name]

        if subject_list:
            filtered_data = [record for record in filtered_data if record["subject_name"] in subject_list]

        if date:
            try:
                datetime.strptime(date, "%d-%m-%Y")  # Validate date format
                filtered_data = [record for record in filtered_data if record["date"] == date]
            except ValueError:
                print("Error: 'date' must be in 'day-month-year' format.")
                return

        if date_range:
            if not isinstance(date_range, tuple) or len(date_range) != 2:
                print("Error: 'date_range' must be a tuple with start and end dates in 'day-month-year' format.")
                return
            try:
                start_date = datetime.strptime(date_range[0], "%d-%m-%Y")
                end_date = datetime.strptime(date_range[1], "%d-%m-%Y")
                filtered_data = [
                    record for record in filtered_data
                    if start_date <= datetime.strptime(record["date"], "%d-%m-%Y") <= end_date
                ]
            except ValueError:
                print("Error: Dates in 'date_range' must be in 'day-month-year' format.")
                return

        if index_range:
            if not isinstance(index_range, tuple) or len(index_range) != 2 or not all(isinstance(i, int) for i in index_range):
                print("Error: 'index_range' must be a tuple of two integers.")
                return
            start_idx, end_idx = index_range
            filtered_data = [record for record in filtered_data if start_idx <= record["index"] <= end_idx]

        reaction_times = [record["reaction_time_ms"] for record in filtered_data]

        if reaction_times:
            count = len(reaction_times)
            min_time = np.round(min(reaction_times), 4)
            max_time = np.round(max(reaction_times), 4)
            avg_time = np.round(stat.mean(reaction_times), 4)
            median_time = np.round(stat.median(reaction_times), 4)
            mode_time = np.round(stat.mode(reaction_times), 4) if len(set(reaction_times)) < len(reaction_times) else None
            variance_time = np.round(stat.variance(reaction_times), 4) if len(reaction_times) > 1 else 0
            std_dev = np.round(stat.stdev(reaction_times), 4) if len(reaction_times) > 1 else 0
            range_time = np.round(max_time - min_time, 4)
            iqr_time = np.round(iqr(reaction_times), 4)
            q1 = np.round(sorted(reaction_times)[int(0.25 * (count - 1))], 4)
            q3 = np.round(sorted(reaction_times)[int(0.75 * (count - 1))], 4)

            return [count, min_time, max_time, avg_time, median_time, mode_time, variance_time, std_dev, range_time, iqr_time, q1, q3]
        else:
            print("No data available for the specified filter criteria.")
            return None

    except FileNotFoundError:
        print("No records found. The JSON file does not exist.")
        return None

def pretty_stats(subject_name=None, subject_list=None, date=None, date_range=None, index_range=None):
    """
    Display statistics in a readable format by calling the stats() function.

    Parameters:
    subject_name (str, optional): Filter by a single subject's name.
    subject_list (list, optional): Filter by a list of subject names.
    date (str, optional): Specific date in "day-month-year" format to filter by.
    date_range (tuple, optional): Tuple with start and end dates in "day-month-year" format to filter by.
    index_range (tuple, optional): Tuple containing start and end indices to filter by.
    """
    stats_result = stats(subject_name, subject_list, date, date_range, index_range)

    if stats_result:
        count, min_time, max_time, avg_time, median_time, mode_time, variance_time, std_dev, range_time, iqr_time, q1, q3 = stats_result
        print("Reaction Time Statistics:")
        print(f"Count: {count} drops")
        print(f"Minimum Reaction Time: {min_time} ms")
        print(f"Maximum Reaction Time: {max_time} ms")
        print(f"Mean Reaction Time: {avg_time} ms")
        print(f"Median Reaction Time: {median_time} ms")
        print(f"Mode: {mode_time if mode_time is not None else 'No mode'} ms")
        print(f"Variance: {variance_time} ms")
        print(f"Standard Deviation: {std_dev} ms")
        print(f"Range: {range_time} ms")
        print(f"Interquartile Range (IQR): {iqr_time} ms")
        print(f"25th Percentile (Q1): {q1} ms")
        print(f"75th Percentile (Q3): {q3} ms")
    else:
        print("No data available for the specified filter criteria.")



def record_ruler_drop(subject_name, distance_in_inches):
    """
    Calculate reaction time and record details including subject name,
    date, distance, and reaction time in a JSON file.

    Parameters:
    subject_name (str): Name of the subject catching the ruler.
    distance_in_inches (float): Distance the ruler fell, in inches.

    Returns:
    float: The reaction time in milliseconds, rounded to 4 decimal places.
    """

    reaction_time_ms = round(record_reaction_time(distance_in_inches), 4)

    try:
        with open("reaction_times.json", "r") as file:
            data = json.load(file)
    except FileNotFoundError:
        data = []

    # Determine the next available index
    if data:
        existing_indices = [record["index"] for record in data]
        next_index = max(existing_indices) + 1
    else:
        next_index = 1

    record = {
        "index": next_index,
        "subject_name": subject_name,
        "date": datetime.now().strftime("%d-%m-%Y"),
        "distance_in_inches": distance_in_inches,
        "reaction_time_ms": reaction_time_ms
    }

    data.append(record)
    with open("reaction_times.json", "w") as file:
        json.dump(data, file, indent=4)

    print(f"Record Created:\nIndex: {next_index}\nSubject: {subject_name}\nDate: {record['date']}\n"
          f"Distance Fallen: {distance_in_inches} inches\nReaction Time: {reaction_time_ms} ms\n")

    return reaction_time_ms
````

## generate_drop_distances
---

Note: For example data, we are using a special random number generator to create an array of realistic drop distances.

````python
import numpy as np
from scipy.stats import truncnorm

def generate_drop_distances(mean, std_dev, low, high, size):
    """
    Parameters:
    mean (float): The center point of the distribution, where values are most frequent.
    std_dev (float): The standard deviation, controlling the spread around the mean.
    low (float): The minimum boundary for the generated values.
    high (float): The maximum boundary for the generated values.
    size (int): The number of random drop distances to generate.

    Returns:
    np.ndarray: An array of generated drop distances.
    """
    a, b = (low - mean) / std_dev, (high - mean) / std_dev
    distribution = truncnorm(a, b, loc=mean, scale=std_dev)

    return np.round(distribution.rvs(size), 2)


# Parameters for the normal distribution
mean = 7.5
std_dev = 1.5  # Controls the spread around the mean
low, high = 2, 12  # Distribution bounds similar to real life (a ruler)

example_drops = generate_drop_distances(mean, std_dev, low, high, 90)
print(example_drops)
````
